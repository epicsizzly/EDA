```python
df = pd.read_csv('H:\Data Science Work\Data science Projects\data analytics\sales data insights\cleaned_world_happiness_indext.csv')
```

```python
df.rename(columns={'Country name': 'Country_name'}, inplace=True)
df.columns
df.head()
```

# feature engineering

```python
# Example: Create a new column 'Mean_Happiness' by averaging happiness scores over years
df['Mean_Happiness'] = df.groupby('Country_name')['Life Ladder'].transform('mean')
```

```python
df['Social_Wellbeing'] = df['Social support'] + df['Healthy life expectancy at birth'] / 10
```

```python
# Example: Create a new column 'Income_Category' based on GDP per capita
bins = [0, 10000, 20000, 30000, float('inf')]
labels = ['Low', 'Medium', 'High', 'Very High']
df['Income_Category'] = pd.cut(df['Log GDP per capita'], bins=bins, labels=labels, right=False)
```

```python
df.head()
```

# correlation

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

# time series analysis

```python
df = df[df['year'] != 2005]
df[df['Country_name'] == 'Canada']['year'].head()
```

```python
import matplotlib.pyplot as plt

mean_happiness_by_year = df.groupby(df['year'])['Life Ladder'].mean()

# Plot the mean happiness score over time
plt.figure(figsize=(12, 6))
mean_happiness_by_year.plot(marker='o', linestyle='-')
plt.title('Mean World Happiness Index Over Time (2008-2019)')
plt.xlabel('Year')
plt.ylabel('Mean Life Ladder (Happiness Score)')
plt.grid(True)
plt.show()
```

# Visualization

```python
# Assuming 'year' is a numeric column (not datetime)
# Create a dictionary to store DataFrames for each year
df_by_year = {}

# Iterate over unique years
for year in df['year'].unique():
    # Filter data for the current year
    df_year = df[df['year'] == year].copy()
    # Store the DataFrame in the dictionary
    df_by_year[year] = df_year

# Now, dfs_by_year contains DataFrames for each year
# Access them like dfs_by_year[2006], dfs_by_year[2007], and so on...
df_by_year[2006]
```

```python
# Sort the DataFrame by 'Life Ladder' in ascending order
asc_df_life_ladder = df.sort_values(by='Life Ladder')
asc_df_life_expectancy = df.sort_values(by='Healthy life expectancy at birth')
asc_df_log_gdp = df.sort_values(by='Log GDP per capita')
asc_df_social_support = df.sort_values(by='Social support')
asc_df_generosity = df.sort_values(by='Generosity')
asc_df_freedom = df.sort_values(by='Freedom to make life choices')
asc_df_positive = df.sort_values(by='Positive affect')
asc_df_corruption = df.sort_values(by='Perceptions of corruption')
asc_df_negative = df.sort_values(by='Negative affect')
```

```python
import matplotlib.pyplot as plt

# Visualize the top 20 countries by Life Ladder
top_20_life_ladder = asc_df_life_ladder.nlargest(20, 'Life Ladder')


plt.bar(top_20_life_ladder['Country_name'], top_20_life_ladder['Life Ladder'])
plt.title('Top 20 Countries by Life Ladder')
plt.xlabel('Country')
plt.ylabel('Life Ladder')
plt.xticks(rotation=45, ha='right')  # Rotate country names for better visibility
plt.tight_layout()
plt.show()
```

```python
# Visualize the top 20 countries by Healthy Life Expectancy
top_20_life_expectancy = asc_df_life_expectancy.nlargest(20, 'Healthy life expectancy at birth')


plt.bar(top_20_life_expectancy['Country_name'], top_20_life_expectancy['Healthy life expectancy at birth'], color='green')
plt.title('Top 20 Countries by Healthy Life Expectancy')
plt.xlabel('Country')
plt.ylabel('Healthy Life Expectancy')
plt.xticks(rotation=45, ha='right')  # Rotate country names for better visibility
plt.tight_layout()
plt.show()
```
